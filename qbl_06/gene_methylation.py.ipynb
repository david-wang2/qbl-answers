```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load reference data
ref = []

with open('mm10_refseq_genes_chr6_50M_60M.bed','r') as f:
    for line in f:
        begin = int(line.split('\t')[4])
        end = int(line.split('\t')[5])
        name = line.split('\t')[12]
        ref.append((begin,end,name))
```

```python
# load e4 data
e4 = {}
count = 0
with open('E4_sorted.bedGraph','r') as f:
    for line in f:
        if (count > 0):
            begin = int(line.split('\t')[1])
            end = int(line.split('\t')[2])
            meth = float(line.split('\t')[3].split('\n')[0])
            if (meth > 0.0):
                e4[begin] = meth/100
        count += 1

e5 = {}
count = 0
with open('E5_sorted.bedGraph','r') as f:
    for line in f:
        if (count > 0):
            begin = int(line.split('\t')[1])
            end = int(line.split('\t')[2])
            meth = float(line.split('\t')[3].split('\n')[0])
            if (meth > 0.0):
                e5[begin] = meth/100
        count += 1
```

```python
gene_meth = {}
prev_begin = []
# Check for methylation for each gene
for i in range(len(ref)):
    begin = ref[i][0]
    end = ref[i][1]
    e4_meth = 0.0
    e5_meth = 0.0
    for loc_e4 in e4.keys():
        if (loc_e4 >= begin) and (loc_e4 <= end):
            e4_meth += e4[loc_e4]
    for loc_e5 in e5.keys():
        if (loc_e5 >= begin) and (loc_e5 <= end):
            e5_meth += e5[loc_e5]
    if (e4_meth > 0):   
        gene_meth[ref[i][2]] = (e5_meth/e4_meth)
```

```python
with open("meth_change.dat",'w') as f:
    f.write(str(gene_meth))
```
